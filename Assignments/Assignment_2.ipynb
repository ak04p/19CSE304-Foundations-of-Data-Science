```python
edu_df = pd.read_csv("https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv")
edu_df
```

```python
df2 = edu_df
year = df2['Year'].str.split('/', expand = True)
df2['Year'] = year[2].str.split(' ', expand = True)[0]
df2
```

```python
adult_df = df2.query('Year == "14" & Age == "18 to 64" | Age == "65 to 80+"')
adult_df
```

```python
edu_df.groupby("Educational Attainment")["Population Count"].sum()
```

```python
df5 = adult_df.groupby("Educational Attainment")["Educational Attainment"].count()
print(df5)
df5.count()
```

```python
print((adult_df.groupby("Educational Attainment")["Educational Attainment"].count()/df5.count()))
```

```python
df6 = adult_df.groupby("Educational Attainment")["Population Count"].sum()
print(df6)
df6.count()
print("\n")
print((adult_df.groupby("Educational Attainment")["Population Count"].sum()/df6.sum())*100)
```
